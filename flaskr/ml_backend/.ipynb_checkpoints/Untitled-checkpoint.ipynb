```python
import numpy as np
from sklearn.cluster import KMeans
import pandas as pd
from sklearn import preprocessing
import itertools
from movielens import *
from sklearn.model_selection import train_test_split
```

```python
ratings = []
item_list = []
user_list = []
d = Dataset()

new_items = []
new_users = []
new_ratings = []

d.load_items("data/u.item", item_list)

d.load_ratings("data/u.base", ratings)

d.load_users("data/u.user", user_list)

d.load_new_items("new_data/ml-latest-small/movies.csv", new_items)

d.load_new_users("new_data/ml-latest-small/ratings.csv", new_users)

d.load_ratings("new_data/ml-latest-small/ratings.csv", new_ratings)

items = pd.read_csv("data/u.item", sep="|", encoding="latin-1", header=None)

users = pd.read_csv("data/u.user", sep="|", encoding="latin-1", header=None)

rating = pd.read_csv("data/u.base", sep='\t', encoding="latin-1", header=None)

test = pd.read_csv("data/u.test", sep='\t', encoding="latin-1", header=None)

updated_rating = pd.read_csv("new_data/ml-latest-small/ratings.csv", sep=",", encoding="latin-1", header=None)

new_items = pd.read_csv("new_data/ml-latest-small/movies.csv", sep=",", encoding="latin-1")




```

```python
base, test = train_test_split(updated_rating, test_size=0.1, random_state=42)

trainDf = pd.DataFrame.from_records(base)
testDf = pd.DataFrame.from_records(test)

trainDf.to_csv(r'new_data/ml-latest-small/train.csv')
testDf.to_csv(r'new_data/ml-latest-small/test.csv')
```

```python


users[2] = le.fit_transform(users[2])
users[3] = le.fit_transform(users[3])
users[4] = le.fit_transform(users[4])

items = items.drop([1,2,3,4], axis=1)


```

```python
X = items.values
Y = users.values
```

```python
kmeans_x = KMeans(n_clusters=50, random_state=0).fit(X)
kmeans_y = KMeans(n_clusters=30, random_state=0).fit(Y)
```

```python
item_labels = {}
item_clusters = {}
for (cluster, item) in itertools.zip_longest(kmeans_x.labels_, item_list):
    item_labels[item.id] = cluster
    if cluster not in item_clusters.keys():
        item_clusters[cluster] = []
    item_clusters[cluster].append(item)
    
user_labels = {}
user_clusters= {}
for(cluster, user) in itertools.zip_longest(kmeans_y.labels_, user_list):
    user_labels[user.id] = cluster
    if cluster not in user_clusters.keys():
        user_clusters[cluster] = []
    user_clusters[cluster].append(user)

    
    
user_clusters[0]
```

```python
rating_clusters= {}
for r in ratings:
    cluster_id = item_clusters[r.item_id]
    if cluster_id not in rating_clusters.keys():
        rating_clusters[cluster_id] = {}
    if r.user_id not in rating_clusters[cluster_id].keys():
        rating_clusters[cluster_id][r.user_id] = []
    rating_clusters[cluster_id][r.user_id].append(r)
```

```python
new_ratings= []
for key, value in rating_clusters.items():
    for key1, value1 in rating_clusters[key].items():
        avg = 0
        for r in value1:
            avg += r.rating
        avg = avg/len(value1)
        rating_clusters[key][key1] = avg
rating_clusters[0]
```
